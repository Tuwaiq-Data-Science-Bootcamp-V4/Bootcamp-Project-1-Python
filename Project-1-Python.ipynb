## Personal Finance Manager

### Overview
This is a simple personal finance manager program that allows you to track your monthly savings progress towards a specified savings goal. It also allows you to view your savings history and total savings amount.

### How to Use
1. Run the program.
2. Enter your monthly salary and savings goal.
3. Choose from the following options:
   - 1. Add Monthly Savings: Enter your monthly savings amount to update your total savings.
   - 2. View Savings Progress: See the percentage of your savings goal achieved.
   - 3. View Savings History: View the history of your monthly savings.
   - 4. View Total Savings: View the total savings amount.
   - 5. Exit: Quit the program.

### Data Types Used
- Dictionary: `user_info` is used to store the user's salary and savings goal.
- List: `savings_history` is used to store the history of monthly savings amounts.

```python
def calculate_savings_progress(total_savings, savings_goal):
    return (total_savings / savings_goal) * 100

def calculate_total_savings(savings_history):
    return sum(savings_history)

def main():
    user_info = {}  # Dictionary to store user information
    user_info['salary'] = float(input("Enter your monthly salary: "))
    user_info['savings_goal'] = float(input("Enter your savings goal: "))
    
    total_savings = 0.0
    savings_history = []
    
    while True:
        print("Options:")
        print("1. Add Monthly Savings")
        print("2. View Savings Progress")
        print("3. View Savings History")
        print("4. View Total Savings")
        print("5. Exit\n")
        
        option = int(input("Enter your choice: "))
        
        if option == 1:
            monthly_savings = float(input("Enter your monthly savings amount: "))
            total_savings += monthly_savings
            savings_history.append(monthly_savings)
            print("Monthly savings added successfully.\n")
            
        elif option == 2:
            progress = lambda x, y: (x / y) * 100
            print("You have achieved {:.2f}% of your savings goal.\n".format(progress(total_savings, user_info['savings_goal'])))
            
            
        elif option == 3:
            print("Savings History:")
            for index, savings in enumerate(savings_history, 1):
                print("Month {}: {}".format(index, savings))
            print("\n")
                
        elif option == 4:
            total = calculate_total_savings(savings_history)
            print("Total Savings: {}\n".format(total))
            
        elif option == 5:
            progress = calculate_savings_progress(total_savings, user_info['savings_goal'])
            print("Exiting the program.")
            print("You have achieved {:.2f}% of your savings goal.".format(progress))
            print("Total Savings: {}".format(total_savings))
            break
            
        else:
            print("Invalid option. Please try again.")

if __name__ == "__main__":
    main()
```
